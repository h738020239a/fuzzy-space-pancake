```python
import pandas as pd
import seaborn as sb
```

```python
dataframe = pd.read_csv('data.csv')
```

# Group (Months) by (gp, unknown) then (count them)

```python
dataframe = dataframe.groupby(['month','type']).agg(count=('type', 'count'))
print(dataframe)
```

# Print as single(s)

```python
# could shorten to df but for ease of read
dataframe = dataframe.reset_index()
print(dataframe)
```

```python
# https://seaborn.pydata.org/tutorial/color_palettes.html
sb.set_palette("PuBuGn_d")
sb.lineplot(x="month", y="count", hue='type', data=dataframe)
```
